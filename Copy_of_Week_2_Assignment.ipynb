1. Create a student grade management system that stores student information and calculates statistics.

**Input Format:**

First line: number of students

Next lines: student_name subject1:grade1,subject2:grade2,...


**Output Format:**

Line 1: Dictionary of all student records

Line 2: List of students sorted by average grade (descending)

Line 3: Tuple of (highest_average, lowest_average, class_average)


**Example:**

**Input:**

3

Alice Math:85,Science:90,English:78

Bob Math:92,Science:88,English:95

Charlie Math:76,Science:82,English:80

**Output:**

{'Alice': {'Math': 85, 'Science': 90, 'English': 78}, 'Bob': {'Math': 92, 'Science': 88, 'English': 95}, 'Charlie': {'Math': 76, 'Science': 82, 'English': 80}}

['Bob', 'Alice', 'Charlie']

(91.67, 79.33, 84.44)

```python
n = int(input())
i = 0
students = {}
percent = {}

while True:
  if i < n:
    str = input()
    str = str.replace(':',' ').replace(',',' ')
    str = str.split()
    name = str[0]
    math = int(str[2])
    sci = int(str[4])
    eng = int(str[6])

    students[name] = {'Math': math, 'Science': sci, 'English': eng}

    percent[name] = (math + sci + eng) / 3
    i += 1
  else:
    break

print(students)

sorted_percent = sorted(percent.items(), key = lambda x: x[1], reverse = True)

print(list(item[0] for item in sorted_percent))

avg = (int(sorted_percent[0][1]) + int(sorted_percent[-1][1])) / 2

print(((sorted_percent[0][1], sorted_percent[-1][1], avg)))
```

4. Manage a store inventory with operations like adding stock, processing orders, and generating reports.

**Input Format:**

First line: number of initial inventory items

Next lines: item_name:quantity:price

Following line: number of operations

Next lines: operation_type:parameters


Operations:

ADD:item_name:quantity:price

SELL:item_name:quantity

RESTOCK:item_name:quantity

CHECK:item_name


**Output Format:**

Line 1: Final inventory dictionary

Line 2: List of items that need restocking (quantity < 5)

Line 3: Set of items that were sold

Line 4: Tuple of (total_items, total_value, most_expensive_item)


**Example:**

**Input:**

3

Laptop:10:1000

Mouse:25:20

Keyboard:15:50

4

SELL:Laptop:2

ADD:Monitor:5:300

RESTOCK:Mouse:10

SELL:Keyboard:5

**Output:**

{'Laptop': (8, 1000), 'Mouse': (35, 20), 'Keyboard': (10, 50), 'Monitor': (5, 300)}

['Monitor']

{'Laptop', 'Keyboard'}

(4, 10800, 'Laptop')

```python
n = int(input())
inv = {}
sold_items = set()
for _ in range(n):

  str = input().split(':')
  name = str[0]
  qty = int(str[1])
  price = int(str[2])
  inv[name] = (qty, price)

m = int(input())
sold_items = set()
restock_items = []
for _ in range(m):
   op = input().replace(':',' ')
   op = op.split()
   if op[0] == "ADD":
       item = op[1]
       add_qty = int(op[2])
       pr = int(op[3])
       inv[item] = (add_qty, pr)
       restock_items.append(item)

   elif op[0] == "SELL":
        item = op[1]
        sold_qty = int(op[2])
        pr = int(op[2])
        inv[item] = (int(inv[item][0]) - sold_qty, pr)
        sold_items.add(item)

   elif op[0] == "RESTOCK":
        item = op[1]
        r_qty = int(op[2])
        inv[item] = (int(inv[item][0]) + r_qty, int(inv[item][1]))

   elif op[0] == "CHECK":
        item = op[1]
        if item in inv:
          print(inv[item])
        else:
          print("item not found")



total_item = len(inv)
total_value = 0

for qty, price in inv.values():
  total_value = total_value + (qty*price)

most_expensive = max(inv, key = lambda x:inv[x][1])

print(inv)
print(restock_items)
print(sold_items)
print((total_item, total_value, most_expensive))
```

6. Build a recommendation system that analyzes purchase history and suggests products based on similarity and popularity.

**Input Format:**

First line: number of customers

Next lines: customer_id:purchased_items (comma-separated)

Following line: number of products with categories

Next lines: product_id:category:price

Last line: customer_id to generate recommendations for


**Output Format:**

Line 1: Dictionary of customer purchase profiles

Line 2: Dictionary of product categories with their average prices

Line 3: Set of customers with similar purchase patterns (Jaccard similarity > 0.3)

Line 4: List of top 5 recommended products (not already purchased)

Line 5: Tuple of (total_customers, most_popular_category, recommendation_confidence)


**Example:**

**Input:**

4

C001:P001,P002,P003

C002:P001,P004,P005

C003:P002,P003,P006

C004:P001,P002,P007

5

P001:Electronics:100

P002:Electronics:150

P003:Books:20

P004:Electronics:200

P005:Books:15

C001


**Output:**

{'C001': {'Electronics', 'Books'}, 'C002': {'Electronics', 'Books'}, 'C003': {'Electronics', 'Books'}, 'C004': {'Electronics'}}

{'Electronics': 137.5, 'Books': 17.5}

{'C002', 'C003', 'C004'}

['P004', 'P005']

(4, 'Electronics', 0.75)

```python
n = int(input())
customers = {}

for _ in range(n):
  cid, items = input().split(':')
  customers[cid] = set(items.split(','))

m = int(input())

products = {}
category_prices = {}

for _ in range(m):
  pid, category, price = input().split(':')
  price = int(price)
  products[pid] = (category, price)
  if category not in category_prices:
    category_prices[category] = []
    category_prices[category].append(price)

target = input()

profiles = {}

for cid, items in customers.items():
  categories = set()

  for item in items():
    if item in products:
      categories.add(products[item][0])
  profiles[cid] = categories

print(profiles)

avg_prices = {}

for cat, prices in category_prices.items():
  avg_prices[cat] = sum(prices)/len(prices)

print(avg_prices)

similar = set()

target_items = customers[target]

for cid, items in customers.items():
  if cid == target:
    continue

  similarity = len(target_items & items)/len(target_items|items)

  if similarity > 0.3:
    similar.add(cid)

print(similar)

recommend = set()

for cid in similar:
  recommend.update(customers[cid] - target_items)

print(list(recommend)[:5])

category_count = {}

for cats in profiles.values():
   for cat in cats:
    category_count = category_count.get(cat, 0) + 1

popular = max(category_count, key = category_count.get)

confidence = len(similar)/n

print((n, popular, round(confidence, 2)))


```

7. Manage a tournament bracket system with player rankings, match results, and statistics.

**Input Format:**

First line: number of players

Next lines: player_name:initial_rating

Following line: number of matches played

Next lines: player1:player2:winner:score1:score2


**Output Format:**

Line 1: Dictionary of player statistics (wins, losses, rating)

Line 2: List of players ranked by final rating

Line 3: Set of players who beat higher-rated opponents

Line 4: Tuple of matches played by each player (as tuple pairs)

Line 5: Tournament statistics (total_matches, upsets, average_rating_change)


**Example:**

**Input:**

4

Alice:1200

Bob:1100

Charlie:1300

David:1000

3

Alice:Bob:Alice:3:1

Charlie:David:Charlie:2:0

Alice:Charlie:Charlie:1:2


**Output:**

{'Alice': {'wins': 1, 'losses': 1, 'rating': 1190}, 'Bob': {'wins': 0, 'losses': 1, 'rating': 1090}, 'Charlie': {'wins': 2, 'losses': 0, 'rating':
1320}, 'David': {'wins': 0, 'losses': 1, 'rating': 990}}

['Charlie', 'Alice', 'Bob', 'David']

set()

[('Alice', 2), ('Bob', 1), ('Charlie', 2), ('David', 1)]

(3, 0, 15.0)

```python
n = int(input())
pl = {}
p = []
for _ in range(n):
  name, rating = input().split(':')
  rating = int(rating)
  pl[name] = {'wins': 0, 'losses': 0, 'rating': rating}
  p.append(name)

m = int(input())
up = set()
matches_played = {}
for _ in range(m):
  p1, p2, win, s1, s2 = input().split(':')
  s1 = int(s1)
  s2 = int(s2)
  loss = p2 if win == p1 else p1

  pl[win]['wins'] += 1
  pl[loss]['losses'] += 1

  pl[win]['rating'] += 10
  pl[loss]['rating'] -= 10

  matches_played[p1] = matches_played.get(p1, 0) + 1
  matches_played[p2] = matches_played.get(p2, 0) + 1

  old1 = pl[p1]["ratings"]
  old2 = pl[p2]["ratings"]

  if p1 == win and old1 < old2:
    up.add((p1, p2))
  elif p2 == win and old2 < old1:
    up.add((p2, p1))

print(pl)

rankings = sorted(pl, key = lambda x:pl[x]["ratings"], reverse = True)

print(rankings)

print(up)

play = [(name, matches_played.get(name, 0)) for name in pl]

print(play)
avg = (m*20)/n

print((m, len(up), round(avg, 2)))
```

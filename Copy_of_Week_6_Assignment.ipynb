### Question 1
Write a function that returns the frequency of each character in a randomly generated string of 100 letters.

```python
import random
import string
from collections import Counter

n = 100
letter = "".join(random.choices(string.ascii_lowercase, k = n))

def count1(string):
  freq = Counter(string)
  print(freq)

count1(letter)

```

### Question 3
Create a list of the first 10 factorial numbers and return their sum.

```python
from math import factorial

def fact_sum():
  arr = []

  for i in range(10):
    arr.append(factorial(i))

  return sum(arr)

print(fact_sum())
```

### Question 4

### Question 7
Compute the cumulative product of a list of 10 random integers.

```python
import random

arr = [random.randint(1, 100) for _ in range(10)]

def cp(arr):
  result = []

  for i in range(10):
    if i == 0:
      result.append(arr[i])
    else:
      result.append(arr[i]*result[i - 1])

  return result

print(cp(arr))


```

### Question 8
Generate all 2-combinations from a given list of numbers and return only those whose sum is greater than 10.

```python
from itertools import permutations

def find_pairs(arr):
  pairs = permutations(arr, 2)

  for pair in pairs:
    if __builtins__.sum(pair) > 10:
      print(pair)

c = list(map(int, input().split()))

find_pairs(c)
```

### Question 10
Simulate a leaderboard system where scores are inserted dynamically, and the top 5 scores can be retrieved at any point.

```python
arr = []

def insert_score(score):
  while True:
    if score == "":
      break

    else:
      arr.append(int(score))
      arr.sort(reverse = True)

      if len(arr) < 6:
         print(arr)

      else:
         print(arr[:5])

    score = input()

score = input()
insert_score(score)
```

### Question 11
Create a function that calculates the average of the squares of numbers divisible by 3 from a random list of 50 numbers.

```python
import random

arr = [random.randint(1, 100) for _ in range(50)]

def asnd(arr):
  b = []
  for i in range(50):
    if arr[i] % 3 == 0:
      b.append(arr[i]**2)

  return sum(b)/len(b)

print(asnd(arr))
```

### Question 12
Write a program that continuously prints values from a list in a repeating pattern (infinite loop simulation, but limit to 15 iterations).

```python
from itertools import cycle, islice

def repeat_list(arr):

  for x in islice(cycle(arr), 15):
    print(x)

c = list(map(int, input().split()))

repeat_list(c)

```

### Question 13
Implement a custom reduce function that returns the greatest common divisor (GCD) of all numbers in a list.

```python
from math import gcd

def custom_reduce(arr):
  for i in range(len(arr) - 1):
    g = gcd(arr[i], arr[i + 1])

  return g


c = list(map(int, input().split()))

print(custom_reduce(c))
```

### Question 15
Create a function to check if a randomly generated 8-character string is made up entirely of punctuation characters.

```python
import random
import string

characters = string.printable
word = "".join(random.choices(characters, k = 8))

def check_p(word):
  for i in word:
    if i in string.punctuation:
      return True
    else:
      return False

print(check_p(word))
```

### Question 17
Create a custom filter that removes every second element from a sequence until the list is empty.

```python
def filt(arr):
  result = []
  if len(arr) <= 1:
     return []
  else:
    for i in range(len(arr)):
      if i % 2 == 0:
         result.append(arr[i])

    return filt(result)

c = list(map(int, input().split()))

print(filt(c))
```

### Question 18
Generate a list of 20 random floating-point numbers between 1 and 100 and sort them based on the decimal part.

```python
import random

arr = [random.uniform(1, 100) for _ in range(20)]

def sort_decimal(arr):
  arr.sort(key = lambda x: x % 1)

  return arr

print(sort_decimal(arr))
```

### Question 19
Compute how many unique ways you can form 3-digit numbers using the digits 1 through 5 without repeating digits.

```python
from itertools import combinations

arr = [1, 2, 3, 4, 5]

def unique_way(arr):
  c = list(combinations(arr, 3))

  return len(c)

print(unique_way(arr))
```

### Question 20
Build a basic expression evaluator that computes results from a list of (operation, operand1, operand2) tuples.

```python
import operator

ops = {
    '+': operator.add,
    '-': operator.sub,
    '*': operator.mul,
    '/': operator.truediv,
    '%': operator.mod,
    '^': operator.xor,
    '&': operator.and_,
    '|': operator.or_,
    '~': operator.not_,
    '<<': operator.lshift,
    '>>': operator.rshift,
    }


def op_eval(exp):
  return[ops[op](a, b) for op, a, b in exp]


exp = eval(input())
print(op_eval(exp))


```

**2.**Given a string, print the number of vowels in it.

**Input Format:**

A single string (lowercase only).

**Output Format:**

A single integer — count of vowels.

```python
str = input("Enter a lowercase string: ")
a = len(str)
i = 0
count = 0
while True:
  if i < a:
    if str[i] == 'a' or str[i] == 'e' or str[i] == 'i' or str[i] == 'o' or str[i] == 'u':
         count = count + 1
         i = i + 1
    else:
         i = i + 1

  else:
     break

print(count)
```

**3.**Given a lowercase character c and a non-negative integer k, output the character that is k positions after c in the alphabet. Wrap around from 'z' to 'a' if needed.

**Input Format:**

One character and one integer separated by space

**Output Format:**

One character (the shifted result)

```python
arr = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
char = input("Enter a lowercase character: ")
num = int(input("Enter a positive number: "))

b = arr.index(char)

print(arr[(num + b)%26])
```

**4.**Keep taking integer input until you get a negative number. Then print the total number of digits (0–9) you encountered in the input before the negative number.

**Input Format:**

One integer per line until a negative number is entered.

**Output Format:**

A single integer (total count of digits seen).

```python
count = 0

a = int(input("Enter integers: "))

while True:
  if a > 0:
    count = count + 1
    a = int(input())
  else:
    break

print(count)
```

# Week 5 Assignment – Dynamic Programming

### 1. Bunny Hops
**Problem:**
A bunny can jump either 1 or 3 steps at a time. Given `n` steps, in how many distinct ways can the bunny reach the top?

**Input:**
An integer `n` (1 ≤ n ≤ 30)

**Output:**
An integer – total number of distinct ways.

```python
def hops(n):
  dp = [0]*(n + 1)
  dp[0] = 1

  for i in range(1, n + 1):

    if i - 1 >= 0:
      dp[i] += dp[i - 1]

    if i - 3 >= 0:
      dp[i] += dp[i - 3]

  return dp[n]

  a = int(input("Enter no of steps: "))
  print("Number of possible ways", hops(a))
```

### 2. Minimum Sum Path in a Grid
**Problem:**
You are given a 2D grid of size `m x n` filled with positive integers. Start from the top-left cell and move only **right** or **down**. Find the **minimum path sum** to reach the bottom-right cell.

**Input:**
- A 2D grid `grid[m][n]`

**Output:**
- Minimum path sum from top-left to bottom-right

```python
def fmc(c):
  m = len(c)
  n = len(c[0])

  dp = [[0]*n for _ in range(m)]

  dp[0][0] = c[0][0]

  for j in range(1, n):
    dp[0][j] = dp[0][j - 1] + c[0][j]

  for i in range(1, m):
    dp[i][0] = dp[i - 1][0] + c[i][0]

  for i in range(1, m):
    for j in range(1, n):
      dp[i][j] = min(dp[i - 1][j], dp[i][j - 1], dp[i - 1][j - 1]) + c[i][j]

  return dp[m - 1][n - 1]

r = int(input("Enter no of rows: "))
c = int(input("Enter no of columns: "))
arr = []

for _ in range(r):
  a = list(map(int, input().split()))
  arr.append(a)

print(fmc(arr))
```

### 3. Robot Paths in an Obstacle Grid
**Problem:**
A robot is at the top-left corner of a `m x n` grid. Some cells are blocked. The robot can move only **right** or **down**. Find the number of ways to reach the bottom-right corner.

**Input:**
- 2D binary matrix `grid[m][n]` where `1` = obstacle, `0` = free

**Output:**
- Integer: total number of valid paths

```python
def path(c):
  m = len(c)
  n = len(c[0])
  obstacle = False
  j = 0
  p_no = 0

  for i in range(1, m):
    j = j + 1
    k = j
    if c[i][k] == 1 and k < n:
      obstacle == True
      k = k + 1
      if c[i][k] == 1
         print("no valid path!")
         break
      else:
        obstacle == False

    if obstacle == False:
      p_no += 1

  return p_no

  r = int(input("Enter no of rows: "))
  c = int(input("Enter no of columns: "))

  for _ in range(r):
    a = list(map(int, input().split()))
    arr.append(a)

  print(path(arr))
```

### 4. Maximum Points Without Adjacent Pick
**Problem:**
You are given an array of `n` integers. Pick elements such that no two picked elements are adjacent. Maximize the sum of picked elements.

**Input:**
- Array `arr[]` of size `n`

**Output:**
- Maximum sum possible

```python
def max_sum(arr):
  m = len(arr)
  dp = [[0 for _ in range(m)] for _ in range(m)]

  for i in range(m):
    for j in range(m):
        dp[i][j] = arr[i] + arr[j]

  m_sum = []

  for i in range(m):
    for j in range(m):
      if i != j:
        if i - j > 1 or j - i > 1:
           m_sum.append(dp[i][j])

  return max(m_sum)

  a = list(map(int, input("Enter the elements of the array: ").split()))
  print(max_sum(a))

```

### 5. Split Array into K Equal-Sum Parts
**Problem:**
Given an array of `n` integers, determine whether it is possible to split it into **k** non-empty subsets such that the **sum of each subset is the same**.

**Input:**
- Array `arr[]` of size `n`, integer `k`

**Output:**
- Boolean: `True` if possible, else `False`

```python
def cp(nums, k):
  total = sum(nums)

  if total % k != 0
      return False

  target = total//k
  nums.sort(reverse = True)
  used = [False]*len(nums)

  def bt(start, curr_sum, count):
    if count == k - 1
       return True

    if curr_sum == target:
      return bt(0, 0, count + 1)

    for i in range(start, len(nums)):
      if not used[i] and curr_sum + nums[i] <= target:
        used[i] = True
        if bt(i + 1, curr_sum + nums[i], count):
          return True
        used[i] = False

    return False

  return bt(0, 0, 0)

  arr = list(map(int, input("Enter array: ").split()))
  k = int(input("Enter k: "))
  print(cp(arr, k))
```

### 6. Longest Alternating Increasing-Then-Decreasing Subsequence
**Problem:**
Given an array of integers, find the **length of the longest subsequence** which first strictly increases and then strictly decreases.

**Input:**
- Array `arr[]` of size `n`

**Output:**
- Length of the longest such subsequence

```python
def lb(arr):
  n = len(arr)

  lis = [1]*n
  lds = [1]*n

  for i in range(n):
    for j in range(i):
      if arr[j] < arr[i]:
        lis[i] = max(lis[i], lis[j] + 1)

  for i in range(n - 1, -1 , -1):
    for j in range(n - 1, i, -1):
      if arr[j] < arr[i]:
        lds[i] = max(lds[i], lds[j] + 1)

  ans = 0
  for i in range(n):
    ans = max(ans, lis[i] + lds[i] - 1)

  return ans

a = list(map(int, input("Enter the elements of the array: ").split()))
print(lb(a))
```

**1.** Linear Search in a List

Write a function to perform linear search on a list of integers.

**Input:** A list of n integers and a target integer x

**Output:** Index of x if found, else -1

```python
def target(a, x):
   if x in a:
      return a.index(x)
   else:
      return -1

arr = list(input("Enter elements: ").split())
x = input("Enter target: ")
print(target(arr, x))


```

**3.** Binary Search (Recursive Implementation)

Write a recursive binary search function for a sorted list.

**Input:** Sorted list of integers and a target value

**Output:** Index of the target or -1 if not found

```python
def search(arr, a, b, x):
  mid = (a + b)//2
  if x < arr[mid]:
    return(arr, a, mid - 1, x)

  if x > arr[mid]:
    return(arr, mid + 1, b, x)

  if arr[mid] == x:
    return mid


p = list(map(int, input("Enter elements: ").split()))
k = int(input("Enter Target: "))

print(search(p, 0, len(p)-1, k)) # error

```

**4.** Insertion Sort Implementation

Implement the insertion sort algorithm without using built-in sort.

**Input:** Unsorted list of integers

**Output:** Sorted list

```python
def sort(arr):
  for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and key < arr[j]:
      arr[j+1] = arr[j]
      j -= 1
    arr[j+1] = key
  return arr

b = list(map(int, input().split()))

print(sort(b))
```

**5.** Merge Sort using Recursion

Implement merge sort using recursion.

**Input:** A list of integers

**Output:** Sorted list

```python
def ms(arr):
  if len(arr) <= 1:
     return arr

  mid = len(arr)//2
  left = arr[:mid]
  right = arr[mid:]

  left = sorted(left)
  right = sorted(right)

  return merge(left, right)

def merge(left, right):
    result = []
    i = 0
    j = 0

    while i < len(left) and j < len(right):
      if left[i] < right[j]:
        result.append(left[i])
        i += 1
      else:
        result.append(right[j])
        j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

a = list(map(int, input().split()))

print(ms(a))
```

```python

import random

cells = [".",".",".","#","T","M"]

cols = random.randint(1, 10)

maze = []

print("maze: ")

for i in range(cols):
  row = ""

  if i == 0:
      row = "S"+ "".join(random.choice(cells) for _ in range(cols - 1))
      maze.append(row)

  if i == cols - 1:
      row = "".join(random.choice(cells) for i in range(cols - 1)) + "E"
      maze.append(row)

  if i > 0 and i < cols - 1:
      for _ in range(cols):
        row += random.choice(cells)

      maze.append(row)

  print(row)

print("maze :",maze)


def g_n(maze):
  graph = {}
  n = len(maze)
  cost = {'S': 0, 'E': 0, '.': 1, '#': 8, 'M': 5, 'T': 3}

  directions = [ (1, 0), (0, 1), (-1, 0), (0, -1)]

  for i in range(n):
    for j in range(n):
       graph[(i, j)] = []

       for dr, dc in directions:
         nr = i + dr
         nc = j + dc

         if 0 <= nr < len(maze) and 0 <= nc < len(maze[nr]):
           if maze[nr][nc] != '#':
             graph[(i, j)].append((maze[nr][nc], cost[maze[nr][nc]], (nr, nc)))

  return graph


print("graph :",g_n(maze))

graph = g_n(maze)

def next_node(graph, node, visited):
    valid = [
        (symbol, cost, coordinate)
        for symbol, cost, coordinate in
     graph[node]
        if coordinate not in visited
    ]

    if not valid:
      return None

    return min(valid, key = lambda x:x[1])



def short_path(graph):
  path = []

  start = next(iter(graph))
  visited = {start}
  current = start

  while True:
    result = next_node(graph, current, visited)

    if result is None:
      break

    symbol, cost, coordinate = result

    path.append((symbol, coordinate))
    visited.add(coordinate)

    if symbol == 'E':
      break

    current = coordinate



  monsters = 0
  for item in path:
    if item[0] == 'M':
      monsters += 1

  print("monsters: ",monsters)

  return path



print("shortest path :",short_path(graph))





```
